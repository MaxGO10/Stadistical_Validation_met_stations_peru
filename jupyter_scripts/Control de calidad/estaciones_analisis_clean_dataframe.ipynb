```python
import numpy as np
from pykrige.ok import OrdinaryKriging
import matplotlib.pyplot as plt

# Coordenadas de los puntos de datos
x = [-70.593, -70.356, -70.373, -70.346, -70.191, -70.119, -69.966, -69.763, -69.491,
     -69.844, -70.316, -70.44, -70.547, -70.912, -69.814, -69.143, -68.356, -70.783, -68.034,
     -68.187, -71.848, -71.604, -71.066]
y = [-14.881, -14.695, -15.356, -15.639, -14.915, -15.136, -14.779, -15.203, -15.39, -15.616,
     -12.083, -12.39, -14.514, -14.163, -16.051, -15.617, -16.695, -14.019, -16.221, -16.763, -13.761,
     -15.64, -16.184]
# Valores de los puntos de datos
values = [160.5, 125.9, 169.3, 230.1, 180.7,
          165.05, 213.5, 128.6, 137.5,
          158.8, 166.82, 69.8, 109.5, 136.8, 126.9,
          173.7, 134.6, 132.4, 187.8, 119.8,
          85.2, 143.4, 157.0]

# Crear una grilla de coordenadas donde queremos estimar los valores
gridx = np.linspace(min(x) - 0.25, max(x) + 0.25, 100)
gridy = np.linspace(min(y) - 0.25, max(y) + 0.25, 100)


# Probar diferentes modelos de variograma
for variogram_model in ['linear', 'power', 'gaussian', 'spherical', 'exponential']:
    # Realizar la interpolación Kriging
    OK = OrdinaryKriging(x, y, values, variogram_model=variogram_model, verbose=True, enable_plotting=True)
    z, ss = OK.execute('grid', gridx, gridy)

    # Graficar los resultados
    plt.figure(figsize=(10, 8))
    plt.contourf(gridx, gridy, z, cmap='viridis')
    plt.scatter(x, y, c='red', label='Data points')
    plt.colorbar(label='Interpolated values')
    plt.legend()
    plt.title(f'Kriging Interpolation - Variogram Model: {variogram_model}')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.show()
```

```python
import numpy as np
from pykrige.ok import OrdinaryKriging
import matplotlib.pyplot as plt

# Crear algunos datos de ejemplo
# Coordenadas de los puntos de datos
lon = [-70.593, -70.356, -70.373, -70.346, -70.191, -70.119, -69.966, -69.763, -69.491, -69.844,
       -70.316, -70.440, -70.547, -70.912, -69.814, -69.143, -68.356, -70.783, -68.034, -68.187,
       -71.848, -71.604, -71.066]
lat = [-14.881, -14.695, -15.356, -15.639, -14.915, -15.136, -14.779, -15.203, -15.390, -15.616,
       -12.083, -12.390, -14.514, -14.163, -16.051, -15.617, -16.695, -14.019, -16.221, -16.763,
       -13.761, -15.640, -16.184]

# Valores de los puntos de datos (pueden ser precipitación, temperatura, etc.)
# Asegúrate de que los valores sean correctos y estén en el rango esperado.
values = [100, 110, 95, 105, 120, 115, 90, 130, 85, 140, 100, 95, 110, 105, 100, 120, 115, 130, 125, 95, 105, 115, 120]

# Crear una grilla de coordenadas donde queremos estimar los valores
gridx = np.linspace(min(lon)-0.5, max(lon)+0.5, 100)
gridy = np.linspace(min(lat)-0.5, max(lat)+0.5, 100)

# Realizar la interpolación Kriging con un modelo de variograma adecuado
OK = OrdinaryKriging(lon, lat, values, variogram_model='exponential', verbose=False, enable_plotting=False)
z, ss = OK.execute('grid', gridx, gridy)

# Graficar los resultados
plt.figure(figsize=(10, 8))
plt.contourf(gridx, gridy, z, cmap='viridis')
plt.scatter(lon, lat, c='red', label='Data points')
plt.colorbar(label='Interpolated values')
plt.legend()
plt.title('Kriging Interpolation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
